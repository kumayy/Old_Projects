```python
# PREDICTION OF WORLDWIDE GROSS BY USING MOVIE BUDGET
```

```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = (pandas.read_csv("cost_revenue_clean.csv"))
```

```python
data
```

```python
data.describe()
```

```python
x = DataFrame(data, columns = ["production_budget_usd"])
y = DataFrame(data, columns = ["worldwide_gross_usd"])
```

```python
plt.figure(figsize =(15,10))
plt.scatter(x, y, alpha = 0.25)
plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget $")
plt.ylabel("Worldwide Gross $")
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(x, y)
```

slope coeffient

```python
regression.coef_
```

```python
# intercept

regression.intercept_
```

```python
plt.figure(figsize =(15,10))
plt.scatter(x, y, alpha = 0.25)
plt.plot(x, regression.predict(x), color = "grey", linewidth =  4)

plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget $")
plt.ylabel("Worldwide Gross $")
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

```python
regression.score(x,y)
```

```python
budget = float(input("budget of the film: \n(1000 USD)"))
print(f"Budget of the film: {(budget*1000):,}")
sum =(budget *1000* (regression.coef_[0][0]) - (regression.intercept_[0]))
print(f"Estimated worldwide gross is {round(sum):,} USD")
```
